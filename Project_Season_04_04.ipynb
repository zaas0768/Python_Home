### 10. 데이터프레임 groupby 다중집계 (agg)

```python
kopoChSeason.head()
```

```python
sortedData = kopoChSeason.sort_values(['REGIONID','PRODUCT'])
```

```python
sortedData.head()
```

```python
groupKey = ['REGIONID','PRODUCT']  # 컬럼명을 담을 변수를 만든다.
```

```python
groupData = sortedData.groupby(groupKey).agg(['mean','std'])["QTY_NEW"]
# sortedData에 그룹화를 시키고 QTY_NEW에 값을 평균과 표준편차를 구한다음 새로운 컬럼을 만들어라.
```

```python
groupData = groupData.reset_index()  #그룹 데이터를 리셋 시킨다. 시키지 않으면 열이 묶여서 나오게 된다.
groupData.head()
```

### 실습 키 조인으로 mergedData 변수에 담아라 (3-2 ppt.48)

```python
refinedSelloutData.head()
```

```python
groupData.head()
```

```python
mergeKey = ['REGIONID','PRODUCT']
```

```python
mergeResult = pd.merge(groupData, refinedSelloutData, left_on=mergeKey, right_on=mergeKey, how="left")\
[['REGIONID','PRODUCT','YEARWEEK','YEAR','WEEK','QTY','QTY_NEW','QTY_MEAN','QTY_STD']]
# [[]]에 합칠 컬럼명을 쓴다. 왼쪽 부터 순서 대로 쓰면 순서대로 컬럼과 그 내용들이 생긴다.
```

```python
mergeResult.head()
```

```python
# pd. merge(데이터프레임명1, 데이터프레임명2, left_on=‘데이터프레임1 조인 키’
#           right_on=‘데이터프레임2 조인 키’, how = “left”) "left"냐 "inner"냐 하는것
# left join하면 다른 키 내용이 없더라도 데이터 내용이 살아 있지만 inner join 하면 다른 키 내용은 사라진다.
# left는 두 데이터 키 내용이 상관 없이 합쳐지지만 inner면 두 데이터 중에 없는 키 데이터는 사라진다.
```
